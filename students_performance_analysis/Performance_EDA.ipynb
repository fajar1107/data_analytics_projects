### Rename Columns

```python
df.columns = [col.strip().replace(" ", "_").lower() for col in df.columns]

df.head()
```

---

### Univariate Analysis

1. Categorical features

```python
categorical_cols = ['gender', 'race/ethnicity', 'parental_level_of_education', 'lunch', 'test_preparation_course']

for col in categorical_cols:
    plt.figure(figsize=(5, 4))
    sb.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

2. Numerical features

```python
numerical_cols = ['math_score', 'reading_score', 'writing_score']

for col in numerical_cols:
    plt.figure(figsize=(20, 4))
    sb.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.show()
```

---

### Bivariate Analysis

1. Gender vs Scores

```python
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sb.boxplot(data=df, x='gender', y=col)
    plt.title(f'{col} by Gender')
    plt.show()
```

2. Test Preparation vs Scores

```python
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sb.boxplot(data=df, x='test_preparation_course', y=col)
    plt.title(f'{col} by Test Preparation Course')
    plt.show()
```

3. Lunch vs Scores

```python
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sb.boxenplot(data=df, x='lunch', y=col)
    plt.title(f'{col} by Lunch Type')
    plt.show()
```
