## **Project Name** - Hotel Booking Analysis  

**Project Type** - EDA  

**Contribution** - Individual  

**Project By** - K Z Mohammed Fajar 

>**Overview:**  

This project conducts an Exploratory Data Analysis (EDA) on hotel booking data to investigate cancellation trends and uncover underlying patterns. It aims to provide actionable insights that hotels can use to minimize cancellations and enhance revenue.

>**Dataset Description:**  

The dataset includes booking records from a city hotel and a resort hotel. Key features include booking dates, length of stay, number of guests (adults and children), type of meal plan, customer type, and booking channels.

>**EDA Process:**  

The analysis is carried out in the following stages:

1. **Data Exploration & Cleaning:**  

    * Initial inspection of raw data to understand structure and content.

    * Handling of missing values and removal of irrelevant or duplicate entries to prepare the dataset for analysis.

2. **Univariate Analysis:**  

    * Examines individual features (e.g., distribution of cancellation rates, booking lead times, and stay durations) to understand overall patterns.

3. **Bivariate Analysis:**  

    * Investigates relationships between two variables, such as the correlation between booking channel and cancellation rate, or lead time and likelihood of cancellation.

4. **Multivariate Analysis:**  

    * Explores interactions among multiple variables to gain deeper insights into complex patterns driving cancellations.

>**Findings & Recommendations:**  

The analysis reveals key factors contributing to cancellations, such as long lead times, certain distribution channels, and booking during peak periods. Based on these insights, the project suggests practical strategies to reduce cancellations:

* Offer early bird discounts or long-stay promotions to encourage commitment.

* Provide added value (e.g., free breakfast, parking) for direct bookings.

* Monitor high-risk segments and adapt pricing or communication strategies accordingly.
